### Data Cleaning

```python
# Renaming Weather Column to Weather Condition

weather_data.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
```

```python
print(weather_data)
```

```python
# Checking Null Values

null_values = weather_data.isnull().sum()
print(null_values)
```

```python
# Clear Weather_Conditions

clear_weather_records = weather_data[weather_data['Weather_Condition'] == 'clear']
print(clear_weather_records)
```

```python
# The number of times the wind speed was exactly 4 km/hr

wind_speed_4_count = (weather_data['Wind Speed_km/h'] == 4).sum()
print('Number of times wind speed was exactly 4 km/hr:', wind_speed_4_count)
```

```python
# Mean visibility of the dataset

mean_visibility = weather_data['Visibility_km'].mean()
print(mean_visibility)
```

```python
# Number of records where wind speed is greater than 24 km/hr and visibility is equal to 25 km

wind_speed_visibility_count = weather_data[(weather_data['Wind Speed_km/h'] > 24) & (weather_data['Visibility_km'] == 25)].shape[0]
print(wind_speed_visibility_count)
```

```python
# The mean value of each column for each weather condition

weather_data.describe()
```

```python
# Finding instances where weather is clear and relative humidity > 50 or visibility > 40
clear_humidity_visibility = weather_data[(weather_data['Weather_Condition'] == 'Clear') & ((weather_data['Rel Hum_%'] > 50) | (weather_data['Visibility_km'] > 40))]
print(clear_humidity_visibility)
```

```python
# Counting the number of weather conditions that include snow

snow_conditions_count = weather_data[weather_data['Weather_Condition'].str.contains('Snow', na=False)].shape[0]
print('Number of weather conditions that include snow:', snow_conditions_count)
```
